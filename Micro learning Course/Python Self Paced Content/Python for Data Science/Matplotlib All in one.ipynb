```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
# To show a line
%matplotlib inline
```

```python
# Scatter Plot
#Set the random seed for NumPy function to keep the results reproducible
np.random.seed(42)

# Generate a 2 by 100 NumPy Array of random decimals between 0 and 1
r = np.random.rand(2,100)

# Plot the x and y coordinates of the random dots on a scatter plot
plt.scatter(r[0],r[1])

# Show the plot
plt.show()
```

```python
# Multiline plots
# Let's try plotting the temperatures of three different cities with the following code
# Prepare the data series
d = [11,12,13,14,15,16,17]
a1 = [15.3,15.4,12.6,12.7,13.2,12.3,11.4]
a2 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
a3 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]

# Plot the lines for each data series
plt.plot(d,a1)
plt.plot(d,a2)
plt.plot(d,a3)

plt.show()
```

```python
# Scatter Plot to show clusters
# seed the random number generator to keep results reproducible
np.random.seed(123)

# Generate 10 random nos around 2 as x-coordinates of first data series
x0 = np.random.rand(10)+1.5

# Generate the y-coordinates another data series similarly
np.random.seed(321) 
y0 = np.random.rand(10)+2

# Generate the y-coordinates another data series similarly
np.random.seed(456)
x1 = np.random.rand(10)+2.5

np.random.seed(789)
y1 = np.random.rand(10)+2

plt.scatter(x0,y0)
plt.scatter(x1,y1)

plt.show()
```

```python
#Adding trendline over a scatter plot
x = list(range(10))
y = x + np.random.rand(10)-0.5

# Calculate the slope and y-intercept of the trendline
fit = np.polyfit(x,y,1)

# Add the trendline
yfit = [n*fit[0] for n in x] + fit[1]
plt.scatter(x,y)
plt.plot(yfit,'black')

plt.show()
```

```python
#Adding axis labels
''' Let's continue with our example of multi-city temperatures. We shall add plt.xlabel('Date') 
and plt.ylabel('Temperature (C)') to label the axes.'''

d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]

plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)

# Set the limit for each axis
plt.xlim(11,15)
plt.ylim(0,30)

plt.xlabel('Date',size=12, fontweight='semibold')
plt.ylabel('Temperature (C)' , size=12,fontweight='semibold')

plt.show()
```

```python
crop_prod = pd.read_csv('OECD-THND_TONNES.txt', delimiter='\t')
years = crop_prod[crop_prod['Crop']=='SOYBEAN']['Year']
rice = crop_prod[crop_prod['Crop']=='RICE']['Value']
wheat = crop_prod[crop_prod['Crop']=='WHEAT']['Value']
maize = crop_prod[crop_prod['Crop']=='MAIZE']['Value']
soybean = crop_prod[crop_prod['Crop']=='SOYBEAN']['Value']
# Plot the data series
plt.plot(years, rice, label='Rice')
plt.plot(years, wheat, label='Wheat')
plt.plot(years, maize, label='Maize')
plt.plot(years, soybean, label='Soybean')

# Label the x- and y-axes
plt.xlabel('Year',size=12,fontweight='semibold')
plt.ylabel('Thousand tonnes',size=12,fontweight='semibold')

# Add grid
plt.grid(linestyle='--',color='#888888',alpha=0.4)

# Add the title and legend
plt.title('Total OECD crop production in 1995-2016', size=14, fontweight='semibold')
plt.legend()

# Show the figure
plt.show()
```

```python
#Saving plots to a file
y = np.linspace(1,2000)
x = 1.0 / np.sin(y)
plt.plot(x,y,'green')
plt.xlim(-20,20)
plt.ylim(1000,2400)
plt.savefig('123', dpi = 300) 
# dpi means 'Dots per inch'(8 * 12)inches i.e (8 * 300)*(12 * 300)=2400 * 3600 pixels in image
plt.show()
```

```python
#Drawing bar plots with error bars using multivariate data
#To draw a multibar plot to show the performance of an imaginary medicine to treat cancer, by 
# comparing the level of an cancer level before and after treatment of a medicine and placebo 
# as control

# Prepare the data series
import pandas as pd
labels_medicine = ['Medicine(Before)','Medicine(After)']
labels_placebo = ['Placebo(Before)', 'Medicine(After)']
medicine = [2.88,1.42]
placebo = [2.72,2.68]
yerr_medicine = [0.12,0.08]
yerr_placebo = [0.24,0.13]

df = pd.DataFrame([medicine,placebo])
df.columns = ['Before','After']
df.index = ['Medicine','Placebo']

# Plot the bar chart with error bars
df.plot(kind='bar',width=0.4,color=['midnightblue','cornflowerblue'],yerr=[yerr_medicine,yerr_placebo])

plt.title('Effect of Medicines')
plt.xlabel('Condition')
plt.ylabel('[hsCRP] (mg/L)')
plt.xticks(rotation=0) # to keep the xtick labels horizontal
plt.legend(loc=(0.4,0.8))

plt.show()
```
